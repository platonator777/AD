```python
def check_password(password):
    uppercase = "ABCDEFGHIJKLMNOPQRSTUVWXYZ"
    lowercase = "abcdefghijklmnopqrstuvwxyz"
    digits = "0123456789"
    special = "!@#$%&*+"
    allowed_chars = uppercase + lowercase + digits + special
    
    if len(password) < 12:
        return False
    
    for char in password:
        if char not in allowed_chars:
            return False

    has_upper = any(char in uppercase for char in password)
    has_lower = any(char in lowercase for char in password)
    has_digit = any(char in digits for char in password)
    has_special = any(char in special for char in password)
    
    return has_upper and has_lower and has_digit and has_special

n = int(input())
for x in range(n):
    password = input()
    print(check_password(password))
```

```python
data = []
n = int(input())
for x in range(n):
    number = int(input())
    data.append(number)

sorted_data = sorted(data)
n = len(sorted_data)

if n % 2 == 1:
    q2 = sorted_data[n // 2]
    lower_half = sorted_data[:n // 2]
    upper_half = sorted_data[n // 2 + 1:]
else:
    q2 = (sorted_data[n // 2 - 1] + sorted_data[n // 2]) / 2
    lower_half = sorted_data[:n // 2]
    upper_half = sorted_data[n // 2:]

m = len(lower_half)
if m % 2 == 1:
    q1 = lower_half[m // 2]
else:
    q1 = (lower_half[m // 2 - 1] + lower_half[m // 2]) / 2

m = len(upper_half)
if m % 2 == 1:
    q3 = upper_half[m // 2]
else:
    q3 = (upper_half[m // 2 - 1] + upper_half[m // 2]) / 2

iqr = q3 - q1

lower_bound = q1 - 1.5 * iqr
upper_bound = q3 + 1.5 * iqr

outliers = [x for x in data if x < lower_bound or x > upper_bound]

print(len(outliers))
```

```python
participants = input().split()
balances = {name: 0 for name in participants}

N = int(input())
total = 0

for _ in range(N):
    name, cost = input().split()
    cost = int(cost)
    balances[name] += cost
    total += cost

fair_share = total / len(participants)

for name in participants:
    balances[name] = round(balances[name] - fair_share, 2)

debtors = []
creditors = []
for name, bal in balances.items():
    if bal < 0:
        debtors.append([name, -bal])
    elif bal > 0:
        creditors.append([name, bal])

i, j = 0, 0
result = []

while i < len(debtors) and j < len(creditors):
    debtor, debt = debtors[i]
    creditor, credit = creditors[j]

    amount = min(debt, credit)
    result.append((debtor, creditor, amount))

    debt -= amount
    credit -= amount

    if debt == 0:
        i += 1
    else:
        debtors[i][1] = debt

    if credit == 0:
        j += 1
    else:
        creditors[j][1] = credit

print(len(result))
for debtor, creditor, amount in result:
    print(f"{debtor} {creditor} {amount:.2f}")
```

```python
from datetime import datetime
from collections import defaultdict

orders = []
filename = "5_data.txt"
with open(filename, 'r', encoding='utf-8') as file:
    for line_num, line in enumerate(file, 1):
        line = line.strip()
        parts = line.split()
        
        date_str = parts[0]
        order_date = datetime.strptime(date_str, "%d.%m.%Y").date()
        cost = float(parts[-1])
        pizza_name = " ".join(parts[1:-1])
        
        orders.append({
            'date': order_date,
            'pizza': pizza_name,
            'cost': cost
        })

if not orders:
    print("Нет данных для анализа")
else:
    pizza_stats = defaultdict(int)
    date_stats = defaultdict(float)
    total_cost = 0
    
    for order in orders:
        pizza_stats[order['pizza']] += 1
        date_stats[order['date']] += order['cost']
        total_cost += order['cost']
    
    most_expensive = max(orders, key=lambda x: x['cost'])
    avg_cost = total_cost / len(orders)
    
    print(f"\nАнализ данных из файла")
    print("="*50)
    
    print("Статистика по пиццам:")
    sorted_pizzas = sorted(pizza_stats.items(), key=lambda x: x[1], reverse=True)
    for pizza, count in sorted_pizzas:
        print(f"  {pizza}: {count} заказ(ов)")
    print("="*50)
    
    print("Статистика по датам:")
    sorted_dates = sorted(date_stats.items(), key=lambda x: x[0])
    for date, total in sorted_dates:
        print(f"  {date.strftime('%d.%m.%Y')}: {total:.2f} руб.")
    print("="*50)
    
    print(f"Самый дорогой заказ: {most_expensive['cost']:.2f} руб.")
    print(f"Дата: {most_expensive['date'].strftime('%d.%m.%Y')}")
    print(f"Пицца: {most_expensive['pizza']}")
    print("="*50)
    
    print(f"Средняя стоимость заказа: {avg_cost:.2f} руб.")
```

```python
import requests

# Загружаем и обрабатываем данные
mbox = requests.get('http://www.py4e.com/code3/mbox.txt').text
email_counts = {}

for line in mbox.split('\n'):
    if line.startswith('From '):
        email = line.split()[1]
        email_counts[email] = email_counts.get(email, 0) + 1


max_email = max(email_counts, key=email_counts.get)
print(f"Самый активный автор: {max_email}")
print(f"Количество писем: {email_counts[max_email]}")
```
